```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('../data/NCAA_Data.csv')
df.head()
```

```python
average_values = (
    df.groupby("SPORT_NAME").mean(numeric_only=True)
)
average_values = average_values.select_dtypes(include=[np.number])
average_values.dropna(axis=1)
average_values.columns.tolist()
```

```python
new_df = average_values[
    [
        "ACADEMIC_YEAR",
        "MULTIYR_APR_RATE_1000_OFFICIAL",
        "MULTIYR_SQUAD_SIZE",
        "MULTIYR_ELIG_RATE",
        "MULTIYR_RET_RATE",
    ]
]
new_df
```
